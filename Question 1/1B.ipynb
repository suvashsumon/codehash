Take grayscale image of size 512*512 and decrease it intensity level resolution by one bit up to reach its binary format observe its change when displaying in the same window size

```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
def decrease_intensiy_resolution(input_image, bits, max_intensity):
    desired_level = 2**bits
    normalized_image = input_image/max_intensity
    output_image = np.uint8(np.floor(normalized_image*desired_level))

    return output_image
```

```python
# reading original image
original_image = plt.imread('peppers2.tif').copy()
plt.imshow(original_image, cmap='gray')
```

```python
# Define the window size for display
window_size = (1024, 1024)
plt.figure(figsize=(16, 16))

# Display the original image in the first subplot
plt.subplot(4, 2, 1)
plt.imshow(original_image, cmap='gray')
plt.title("Original Image")
#plt.axis('off')

i = 7
index = 2
while i>0:
    decreased_image = decrease_intensiy_resolution(original_image, i, 256.0)
    
    plt.subplot(4, 2, index)
    plt.imshow(decreased_image, cmap='gray')
    plt.title("Quantized by {} bits".format(i))
    #plt.axis('off')
    i-=1
    index+=1

plt.tight_layout()
plt.show()
```
